```python
import json
import pandas as pd
import requests
import numpy as np
 
from pyjstat import pyjstat
from matplotlib import pyplot as plt
import sympy as sp
```

```python
#URL
postURL = "https://data.ssb.no/api/v0/no/table/13156/"
```

```python
#Query
apiQuery = {
  "query": [
    {
      "code": "Region",
      "selection": {
        "filter": "vs:LandetNyn",
        "values": [
          "0N"
        ]
      }
    },
    {
      "code": "ContentsCode",
      "selection": {
        "filter": "item",
        "values": [
          "UtleigdeRom"
        ]
      }
    }
  ],
  "response": {
    "format": "json-stat2"
  }
}
```

```python
# Funksjon for å konvertere uttrekket til 2 dataframes, tar 2 paramtre

def apiToDataframe(postURL, query):

    # postUrl som spørringen skal postes mot
    # Spørringen og endepunktet til API-et kan hentes fra Statistikkbanken.

    res = requests.post(postURL, json=query)
    # legger resultat i ds. DS har i tillegg en del metadata
    ds = pyjstat.Dataset.read(res.text)
    # skriver resultatet til to dataframes
    # først dataframe med tekst
    df = ds.write('dataframe')
    # deretter dataframe med koder
    df_id = ds.write('dataframe', naming='id')
    # returnerer også ds i tilfelle en trenger metadata
    return df, df_id, ds
```

```python
df, df_id, ds = apiToDataframe(postURL, apiQuery)
```

Samfunnsøkonomiske effekter av skatt på overnattinger

```python
def x_d(p,T):
    return 500 - 3.2*(p+T)

def x_s(p,t):
    return -100+4.3*(p-t)
```

```python
p_num=np.linspace(0.1,300,100)

def create_ax():
                fig, ax = plt.subplots()
                ax.set_ylabel('Pris', loc='top')
                ax.set_xlabel('Kvantum', loc='right')
                ax.set(xlim=(220,280))
                ax.set(ylim=(70,90))
                ax.spines['top'].set_color('none')
                ax.spines['right'].set_color('none')
                
                return fig, ax

fig, ax = create_ax()

ax.plot(x_d(p_num, 0),p_num, label='Etterspørsel')

ax.plot(x_s(p_num, 0),p_num, label='Tilbud')
ax.plot(x_s(p_num, 4),p_num, label='Tilbud med avgift')


ax.annotate("", (x_s(82.5,0),82.5), (x_s(86.5,4), 86.5), arrowprops={"arrowstyle":"<->", "color":"gray"})

# Vis likevektskvantum før og etter
ax.text(243.5, 79, "A")
ax.text(236, 83, "B")

# tittel
ax.set_title('Likevekt med en avgift')

plt.xticks([])
plt.yticks([])

ax.legend(bbox_to_anchor=(1.5,0.6));
```

```python
ax.plot(x_d(p_num,4),p_num, label='Etterspørsel med avgift', color='blue')
ax.text(238, 78, "C")

plt.xticks([])
plt.yticks([])

ax.annotate("", (x_d(75,0),75), (x_d(71,4), 71), arrowprops={"arrowstyle":"<->", "color":"gray"})
ax.legend(bbox_to_anchor=(1.6,0.6))
fig
```

```python
# definer symboler i sympy

x,p,t,T=sp.symbols('x p t T', positive=True, real=True)

# betingelse for likevekt

eq_cond=sp.Eq(x_d(p,T),x_s(p,t))
eq_cond
```

```python
# løs for p
p_eq=sp.solve(eq_cond,p)
p_eq[0]
```

```python
# finn tilsvarende kvantum
x_eq=x_s(p_eq[0],t)
x_eq
```

```python
# Likevektspris som funksjon av t og T

p_eq_fn=sp.lambdify(
    (t,T), 
    p_eq[0]
)

p_eq_fn(t,T)
```

```python
# Likevektskvantum som funksjon av t og T

x_eq_fn=sp.lambdify(
    (t,T), 
    x_eq
)

x_eq_fn(t,T)
```

```python
# Kvantum omsatt med avgift på selger på t=4
x_eq_fn(4,0)
```

```python
# Kvantum omsatt med avgift på kjøper på T=4
x_eq_fn(0,4)
```

```python
x_opt = round(x_eq_fn(4,0),2)
x_opt
```

Samfunnsøkonomiske effekter av en turistskatt kan være store og jeg skal prøve å vise de og forklare dem. 
Uten en turistskatt vil markedet for overnattinger følge helt vanlige mikroøkonomiske prinsipper med tilbud og etterspørsel. Prisen på et rom vil avhenge av forholdet mellom tilbud og etterspørsel. Disse kan være påvirket av sesong, kostnader, lokasjon eller konkurranse fra andre.



Om en legger på en turistskatt vil prisen på et rom bli høyere, og det vil senke etterspørselen ettersom folk må betale mer for det samme produktet. For å opprettholde etterspørselen og at folk skal ha lettere for å betale rommet, øker overnattingsvirksomhetene tilbudet slik at de møter den nye etterspørselen for å opprettholde kvantumet så mye som mulig. Dette betyr at skatten blir delt mellom konsument og produsent og sees på som nødvendig for produsentene for som sagt, å opprettholde kvantum og pris så mye som mulig.


Fra figuren kan en tydelig se et dødvektstap som oppstår når avgiften inntrer. Dette fører til at prisen på hotellrom øker, noe som igjen fører til en nedgang i antall hotellrom som selges, fordi noen konsumenter blir priset ut av markedet. Eiere av overnattingsvirksomheter kan også finne på å senke tilbudet av hotellrom på grunn av skatten og derfor bidra til økende økonomisk ineffektivitet og effektivitetstap.


Avgiften vil som sagt prise noen konsumenter ut av markedet, og vi vil åpne fordelingseffekter på hvem som kommer av turister, når og hvor de kommer. Generelt alt som er prissensitivt har størst sjanse for å lide av skatten. Dette kan være lavtlønnede familier, små hotell eller små overnattingsvirksomheter. Disse virksomhetene har kanskje ikke likviditet til å gjøre som andre aktører, nemlig dele skatten mellom seg og konsumentene. De må kanskje være nødt til å be konsumentene om å betale den og kan oppnå enda mer dødvektstap. De vil derfor ha størst sannsynlighet for å bli skjøvet ut av markedet. 

På den andre siden kan skatten gi inntekt til myndighetene som kan bruke dette på å gjøre det enklere for folk å komme seg til stedet ved for eksempel forbedret infrastruktur. Kanskje forbedring av lokaltransport, ryddigere og enklere å oppleve stedet eller tilrettelagt for familier med barnevennlige severdigheter, parker, gjøre stedet tryggere og så videre.

Det er vanskelig å kunne konkret si hvor store effekter en kan forvente, men det er enklere å si hva størrelsen på effektene er avhengige av. Det kan være en rekke faktorer, inkludert skattesatsen, konkurranseevnen til det lokale hotellmarkedet og tilgjengeligheten av andre overnattingsmuligheter i området. Når det gjelder skattesatsen, vil høyere skattesatser sannsynligvis ha en større regressiv effekt på lavinntektsindivider, da de er mer sannsynlig å være prissensitive når det gjelder reiseutgifter. Lavere skattesatser kan derimot ha en mindre regressiv effekt, men kan også generere mindre inntekter for kommunen. 


Konkurranseevnen til det lokale hotellmarkedet vil også spille en rolle i å bestemme størrelsen på fordelingseffektene. Hvis det er mange hoteller i området, kan reisende ha flere alternativer for overnatting, noe som kan bidra til å dempe virkningen av skatten. Men hvis det bare er noen få hoteller i området, kan det være mer sannsynlig at reisende blir priset ut av markedet på grunn av skatten, som kan ha en større regressiv effekt.


Til slutt vil også tilgjengeligheten av andre overnattingsmuligheter i området påvirke størrelsen på fordelingsvirkningene. For eksempel, hvis det er mange alternativer for korttidsleie tilgjengelig (som gjennom Airbnb), kan reisende velge å bo i disse overnattingsstedene i stedet for hotell for å unngå skatten. Dette kan føre til en nedgang i etterspørselen etter hotellrom.
