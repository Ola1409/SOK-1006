```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
import pandas as pd
```

```python
df=pd.read_csv("https://uit-sok-1006-v23.github.io/data/suits_watermelons.csv",delimiter=";")
```

```python
df
```

```python
df=df.dropna()
```

```python
df['price']=np.exp(df['log p'])
df['quantum']=np.exp(df['log q'])
df['harvest']=np.exp(df['log h'])
```

```python
df
```

```python
fig, (ax1,ax2)=plt.subplots(1,2)

ax1.plot(df['Year'], df['price'])
ax1.set_xlabel('Year')
ax1.set_ylabel('Price')

ax2.plot(df['Year'], df['quantum'], color= 'black')
ax2.set_xlabel('Year')
ax2.set_ylabel('Quantum')
```

```python
liste = np.arange(20,105,5)
log_liste = np.log(liste)

df_q = {'Q': liste, 'log q': log_liste}
df_q = pd.DataFrame(df_q)

df_q
```

```python
supply = -2.0+1.7*df_q['log q']
demand = 8.5-0.82*df_q['log q']

df_d_s = {'demand (log p)': demand, 'supply (log p)': supply}

df_d_s = pd.DataFrame(df_d_s)

df_total = pd.concat ([df_q, df_d_s], axis = 1)

df_total['demand (p)'] = np.exp(df_total['demand (log p)'])
df_total['supply (p)'] = np.exp(df_total['supply (log p)'])
```

```python
df_total
```

```python
plt.plot(df_total['Q'], df_total['demand (p)'], label = 'demand')
plt.plot(df_total['Q'], df_total['supply (p)'], label = 'supply')
plt.legend()
```

```python
shock = -2.0+1.7*df_total['log q']+0.4

df_total_shock = {'New supply (log p)': shock}
df_total_shock = pd.DataFrame(df_total_shock)

df_total = pd.concat([df_total, df_total_shock], axis = 1)

df_total['New supply (p)'] = np.exp(df_total['New supply (log p)'])

df_total
```

```python
plt.plot(df_total['Q'], df_total['demand (p)'], label='demand')
plt.plot(df_total['Q'], df_total['supply (p)'], label='supply')
plt.plot(df_total['Q'], df_total['New supply (p)'], label='shock')
plt.xlabel('Quantum')
plt.ylabel('Price')
plt.legend()
```

```python
#Gjr om til symboler som en kan bruke til matematikk
logp, logq = sp.symbols('logp logq')
#Regner ut for supply
eq = sp.Eq(-2.0+1.7*logq, logp)
sp.solve(eq, logq)
```

```python
#Gjr det samme for demand
eq = sp.Eq(8.5+0.82*logp, logq)
sp.solve(eq, logp)
```
